```python
import fSRK_ros as ros
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
print(ros.__doc__)
```

```python
r0 = np.array([1,0,0]) # Initial condition
p = np.array([0.2,0.2,5.7,0.01,2]) # Parameters
t0 = 0 # Initial time
tf = 500 # Final time
nout = 500000 # Number of output points
nps = 5 # Numbe of parameters
```

```python
vals = ros.fsrk(r0,t0,tf,nout,nps,p)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(vals[0][0],vals[0][1],vals[0][2], linewidth = 0.3)
fig.set_size_inches(18, 14) 
plt.show()
```
